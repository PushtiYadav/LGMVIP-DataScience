## KNeighBorsClassifiers

```python
Numerical_data = df.drop('Species',axis=1)
Species_data = df['Species']
x_train, x_test, y_train, y_test = train_test_split(Numerical_data, Species_data, test_size=0.25)

kmeans=KNeighborsClassifier(n_neighbors=3)
kmeans.fit(x_train, y_train)
```

```python
print("Accuracy: ",kmeans.score(x_train,y_train)*100)
```
